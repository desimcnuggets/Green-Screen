```python
import cv2
import numpy as np

cap = cv2.VideoCapture('green_hand.mp4')
imag = cv2.imread('green_hand_background.jpg')
#imag1 = cv2.imread('green_cloak_mask (2).jpg')
#cap1 = cv2.VideoCapture('green_cloak_masks.mp4')

i = 0
new_video = []
```

```python
while(cap.isOpened()):
    print('frames processed: ', i)   
  
    val, frame = cap.read()

    if frame is None:
        print('Frame returned is None')
        break    
        
    frame1 = cv2.cvtColor(frame, cv2.COLOR_RGB2HSV)
    #if (i < 160):
        
    for x in range(0, 640):
        for y in range(0, 360):
            if (30 < frame1[y,x][0] < 70):
                frame1[y,x] = 0
            else:
                frame1[y,x] = 255 
    #cv2.imwrite('green_hand_mask.png', frame1)           
                
    for x in range(0,640):
        for y in range(0, 360):
            if (frame1[y,x][0] == 0):
                frame1[y,x] = imag[y,x]
            elif (frame1[y,x][0] == 255):
                 frame1[y,x] = frame[y,x]
        
        
        
    
    new_video.append(frame1)
        
    i = i + 1

new_video = np.array(new_video) 

writer = cv2.VideoWriter('new_video.mp4', cv2.VideoWriter_fourcc(*'PIM1'), 30, (640, 360), True)

for x in new_video:
    writer.write(x)
cap.release()
writer.release()

```
