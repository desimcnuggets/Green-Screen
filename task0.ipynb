```python
import cv2
import numpy as np
```

```python
cap = cv2.VideoCapture('green_hand.mp4')
```

```python
i = 0
new_video = []
```

```python
while(cap.isOpened()):
    print('frames processed: ', i)
    
  
    val, frame = cap.read()

    if frame is None:
        print('Frame returned is None')
        break
        
    if(i % 10 == 0):
        new_video.append(frame)
    
    i = i + 1
```

```python
new_video = np.array(new_video)
```

```python
writer = cv2.VideoWriter('new_vid.mp4', cv2.VideoWriter_fourcc(*'PIM1'), 30, (640, 320), True)
for x in new_video:
    writer.write(x)
```
